```python
# Liniear Search
def linear_search(arr, target):
    for i in range(len(arr)):
      if arr[i] == target:
        return i
    return -1

# contoh
data = [10, 20, 30, 40, 50]
print(linear_search(data, 30)) # output: 2
```

| Case         | Jumlah Operasi | Kompleksitas |
| ------------ | -------------- | ------------ |
| Best Case    | 1              | O(1)         |
| Average Case | n/2            | O(n)         |
| Worst Case   | n              | O(n)         |




```python
# Binary Search (Iternatif)
def binary_search(arr, target):
  left, right = 0, len(arr) - 1

  while left <= right:
    mid = (left + right) // 2

    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      left = mid + 1
    else:
      right = mid - 1

  return -1

# contoh (data harus terurut)
data = [10, 20, 30, 40, 50]
print(binary_search(data, 40)) #output: 3
```

| Case         | Jumlah Operasi | Kompleksitas |
| ------------ | -------------- | ------------ |
| Best Case    | 1              | O(1)         |
| Average Case | log n          | O(log n)     |
| Worst Case   | log n          | O(log n)     |


```python
# Binary Search (Rekursif)
def binary_search_recursive(arr, left, right, target):
  if left > right:
    return -1

  mid = (left + right ) // 2

  if arr[mid] == target:
    return mid
  elif arr[mid] < target:
    return binary_search_recursive(arr, left, mid + 1, target)
  else:
    return binary_search_recursive(arr, mid - 1, target)

  # contoh
  data = [10, 20, 30, 40, 50]
  print(binary_search_recursive(data , 0, len(data)-1, 20)) #output: 1
```

| Case         | Waktu    |
| ------------ | -------- |
| Best Case    | O(1)     |
| Average Case | O(log n) |
| Worst Case   | O(log n) |


```python
# Jump Search

import math

def jump_search(arr, target):
  n = len(arr)
  step = int(math.sqrt(n))
  prev = 0

  # lompat blok
  while arr[min(step, n) - 1] < target:
    prev = step
    step += int(math.sqrt(n))
    if prev >= n:
      return -1
```

| Case         | Kompleksitas |
| ------------ | ------------ |
| Best Case    | O(1)         |
| Average Case | O(√n)        |
| Worst Case   | O(√n)        |


```python
import math

def jump_search(arr, target):
  n = len(arr)
  if n == 0:
      return -1
  step = int(math.sqrt(n))
  prev = 0

  # lompat blok
  while prev < n and arr[min(step, n) - 1] < target:
    prev = step
    step += int(math.sqrt(n))
    if prev >= n:
      return -1

  # Linear search di blok
  for i in range(prev, min(step, n)):
      if arr[i] == target:
          return i

  return -1

# contoh
data = [10, 20, 30, 40, 50, 60, 70]
print(jump_search(data, 60)) #output: 5
```

| Case         | Kompleksitas |
| ------------ | ------------ |
| Best Case    | O(1)         |
| Average Case | O(√n)        |
| Worst Case   | O(√n)        |
| Space        | O(1)         |


```python
import math

# Interpolation Search

def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1
    while low <= high and target >= arr[low] and target <= arr[high]:
        if low == high:
             if arr[low] == target:

                return low
             return -1
        # rumus interpolasi
        # Cek untuk menghindari ZeroDivisionError jika arr[high] == arr[low]
        if arr[high] == arr[low]:
            if arr[low] == target:
                return low
            else:
                return -1

        pos = low + ((target - arr[low]) * (high - low) //
                     (arr[high] - arr[low]))

        if arr[pos] == target:
            return pos
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1

# contoh
data = [10, 20, 30, 40, 50]
print(interpolation_search(data, 50)) # output: 4
```

| Average Case | O(log log n) |
| Worst Case   | O(n)         |
| Space        | O(1)         |
