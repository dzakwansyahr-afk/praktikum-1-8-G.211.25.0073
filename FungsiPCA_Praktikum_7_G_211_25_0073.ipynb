```python
# Fungsi mencari rata-rata tiap kolom
def mean(data):
    rows = len(data)
    cols = len(data[0])

    avg = []
    for j in range(cols):
        jumlah = 0
        for i in range(rows):
            jumlah += data[i][j]
        avg.append(jumlah / rows)

    return avg


# Fungsi normalisasi data
def center_data(data, avg):
    centered = []

    for row in data:
        temp = []
        for i in range(len(row)):
            temp.append(row[i] - avg[i])
        centered.append(temp)

    return centered


# Fungsi transpose matriks
def transpose(matrix):
    hasil = []

    for j in range(len(matrix[0])):
        baris = []
        for i in range(len(matrix)):
            baris.append(matrix[i][j])
        hasil.append(baris)

    return hasil


# Perkalian matriks
def multiply(A, B):
    hasil = []

    for i in range(len(A)):
        baris = []

        for j in range(len(B[0])):
            total = 0

            for k in range(len(B)):
                total += A[i][k] * B[k][j]

            baris.append(total)

        hasil.append(baris)

    return hasil


# Matriks kovarian
def covariance_matrix(data):
    n = len(data)

    T = transpose(data)
    cov = multiply(T, data)

    for i in range(len(cov)):
        for j in range(len(cov[0])):
            cov[i][j] = cov[i][j] / (n - 1)

    return cov


# Power iteration untuk eigenvector terbesar
def power_iteration(matrix, iterations=100):
    n = len(matrix)
    vector = [1] * n

    for _ in range(iterations):
        new_vector = [0] * n

        for i in range(n):
            for j in range(n):
                new_vector[i] += matrix[i][j] * vector[j]

        norm = sum(x*x for x in new_vector) ** 0.5

        for i in range(n):
            new_vector[i] /= norm

        vector = new_vector

    return vector


# PCA utama
def PCA(data):

    avg = mean(data)
    centered = center_data(data, avg)

    cov = covariance_matrix(centered)

    eigenvector = power_iteration(cov)

    projected = []

    for row in centered:
        nilai = 0
        for i in range(len(row)):
            nilai += row[i] * eigenvector[i]

        projected.append(nilai)

    return projected


# Contoh data
X = [
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9]
]


hasil = PCA(X)

print("Hasil PCA:")
for h in hasil:
    print(h)
```
