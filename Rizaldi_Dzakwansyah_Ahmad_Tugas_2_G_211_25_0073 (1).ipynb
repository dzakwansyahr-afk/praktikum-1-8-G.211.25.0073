mencari nilai maksimum

```python
def cari_maksimum(arr):
    maks = arr[0]

    for i in range(1, len(arr)):
        if arr[i] > maks:
            maks = arr[i]

    return maks


# contoh penggunaan
data = [10, 25, 7, 40, 15]
print("Nilai maksimum:", cari_maksimum(data))
```

mengembalikan nilai elemen

```python
def ambil_elemen(arr, indeks):
    if 0 <= indeks < len(arr):
        return arr[indeks]
    else:
        return "Indeks tidak valid"


# contoh penggunaan
data = [10, 25, 7, 40, 15]
print("Elemen indeks ke-2:", ambil_elemen(data, 2))
```

Tugas Algoritma RED-BLACK TREE

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.parent = None
        self.left = None
        self.right = None
        self.color = 1  # 1 = RED, 0 = BLACK [cite: 56]

class RedBlackTree:
    def __init__(self):
        self.TNULL = Node(0)
        self.TNULL.color = 0 # Daun (NIL) harus hitam [cite: 56]
        self.root = self.TNULL

    def insert(self, key):
        node = Node(key)
        node.parent = None
        node.data = key
        node.left = self.TNULL
        node.right = self.TNULL
        node.color = 1 # Simpul baru berwarna merah [cite: 56]

        # Logika penempatan BST standar
        y = None
        x = self.root
        while x != self.TNULL:
            y = x
            if node.data < x.data:
                x = x.left
            else:
                x = x.right

        node.parent = y
        if y == None:
            self.root = node
        elif node.data < y.data:
            y.left = node
        else:
            y.right = node

        print(f"Berhasil memasukkan {key} ke dalam Tree.")

# Cara menjalankan:
rbt = RedBlackTree()
rbt.insert(10)
rbt.insert(20)

    # (Fungsi insert_fixup dan fungsi lainnya mengikuti logika pseudocode di atas)
```
