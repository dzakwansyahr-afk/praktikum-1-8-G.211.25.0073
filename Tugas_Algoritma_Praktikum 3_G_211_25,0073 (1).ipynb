```python
def algoritma_floydWarshall(w):
  n = len(w)
  dist = [list(baris) for baris in w]

  for k in range(n):
    for i in range(n):
      for j in range(n):
        if dist[i][j] > dist[i][k] + dist[k][j]:
          dist[i][j] = dist[i][k] + dist[k][j]

  return dist

INF = float("inf")

data_matrix = [
  [0, 5, INF, 10],
  [INF, 0, 3, INF],
  [INF, INF, 0, 1],
  [INF, INF, INF, 0]
]

hasil = algoritma_floydWarshall(data_matrix)

print("Jarak terpendek: ")
for baris in hasil:
  print(baris)
```
